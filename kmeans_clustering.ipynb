```python
k = 5
max_iter = 200
loss_hist = []

loss_hist = train(X, k, max_iter, True)
```

```python
nb_k = 15
max_iter = 200
bests_loss = []

for k in range(1, nb_k+1):
    loss_hist = train(X, k, max_iter)
    bests_loss.append(loss_hist[len(loss_hist)-1])
```

```python
plt.plot(range(1, nb_k+1), bests_loss)
plt.title("Optimal number of clusters")
plt.ylabel("Loss")
plt.xlabel("Number of clusters")
plt.show()
```
